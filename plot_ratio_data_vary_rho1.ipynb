```python
import matplotlib.pyplot as plt
import csv
from sklearn import cluster
import numpy as np
import seaborn as sns
sns.set(style="whitegrid")
```

```python
bound_data = []
file_name = 'data_for_graphs/ratio_bound_data.csv'
data_file = open(file_name, 'r')
rdr = csv.reader(data_file)
for row in rdr:
    bound_data.append([float(obs) for obs in row])
```

```python
rho1s = [row[0]/row[2] for row in bound_data]
```

```python
print len(rho1s)
print max(rho1s)
print min(rho1s)
```

```python
rho1s_boundaries = [0.0, 1.0, 2.0]
ratios_inverse_rhos = [[1.0/row[-1] for row in bound_data if (row[0]/row[2]) > rho1s_boundaries[i] and (row[0]/row[2]) <= rho1s_boundaries[i+1]] for i in range(2)]
```

```python
ratios_inverse_rhos[0][:10]
```

```python
bins = [i/200.0 for i in range(200)]
width = 1.0/200.0
fig, ax = plt.subplots()
plt.hist(ratios_inverse_rhos, bins=bins, width=width, edgecolor="none", alpha=0.7, stacked=True, color=['red', 'yellow'], label=['0.0 < rho1 < 1.0', '1.0 < rho1 < 2.0'])
plt.xlabel('Ratios')
plt.ylabel('Frequency')
plt.legend(fontsize=14)
ax.set_title('Actual/Bound', fontsize=20)
plt.show()

bins = [i/200.0 for i in range(200)]
width = 1.0/200.0
fig, ax = plt.subplots()
plt.hist(ratios_inverse_rhos, bins=bins, width=width, edgecolor="none", alpha=0.4, stacked=False, color=['red', 'yellow'], label=['0.0 < rho1 < 1.0', '1.0 < rho1 < 2.0'])
plt.xlabel('Ratios')
plt.ylabel('Frequency')
plt.legend(fontsize=14)
ax.set_title('Actual/Bound', fontsize=20)
plt.show()
```

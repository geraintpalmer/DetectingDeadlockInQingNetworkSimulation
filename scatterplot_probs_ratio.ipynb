```python
import matplotlib.pyplot as plt
import csv
from sklearn import cluster
import numpy as np
import seaborn as sns
import random
sns.set(style="whitegrid")
```

```python
bound_data = []
file_name = 'data_for_graphs/clustered_bound_data_2.csv'
data_file = open(file_name, 'r')
rdr = csv.reader(data_file)
for row in rdr:
    bound_data.append([float(obs) for obs in row])
```

```python
betas = [row[-2] for row in bound_data]
p3_fl = [row[-3] for row in bound_data]
p1_fl = [row[-5] for row in bound_data]
```

```python
sample_data = random.sample(bound_data, 50000)
```

```python
groups = [0, 1, 2]
ratios = [[row[-2] for row in sample_data if row[-1] == g] for g in groups]
p3s = [[row[-3] for row in sample_data if row[-1] == g] for g in groups]
p1s = [[row[-5] for row in sample_data if row[-1] == g] for g in groups]
```

```python
colours = ['darkcyan', 'forestgreen', 'orangered']
labels = ['Group A', 'Group C', 'Group B']
```

```python
fig, ax = plt.subplots()
for i in range(3):
    plt.scatter(p3s[i], ratios[i], marker='o', facecolors='none', alpha=0.6, edgecolors=colours[i])
for i in [0, 2, 1]:
    plt.scatter([], [], color=colours[i], label=labels[i])
plt.xlabel(r"$p_{(-3)}$")
plt.ylabel('Ratios')
ax.set_title('Scatterplot ' + r"$p_{(-3)}$" + ' vs ' + r"$E$", fontsize=20)
ax.set_ylim([0,1])
ax.set_xlim([0,0.7])
plt.legend(fontsize=14, loc='best')
plt.show()

fig, ax = plt.subplots()
for i in range(3):
    plt.scatter(p1s[i], ratios[i], marker='o', facecolors='none', alpha=0.6, edgecolors=colours[i])
for i in [0, 2, 1]:
    plt.scatter([], [], color=colours[i], label=labels[i])
plt.xlabel(r"$p_{(-1)}$")
plt.ylabel('Ratios')
ax.set_title('Scatterplot ' + r"$p_{(-1)}$" + ' vs ' + r"$E$", fontsize=20)
ax.set_ylim([0,1])
ax.set_xlim([0,1])
plt.legend(fontsize=14, loc='upper center')
plt.show()
```

```python
import matplotlib.pyplot as plt
import csv
import seaborn as sns
sns.set(style="whitegrid")
```

```python
bound_data = []
file_name = 'data_for_graphs/ratio_bound_data.csv'
data_file = open(file_name, 'r')
rdr = csv.reader(data_file)
for row in rdr:
    bound_data.append([float(obs) for obs in row])
```

```python
print bound_data[0]
```

```python
ratios = [row[-1] for row in bound_data]
```

```python
ratios_inverse = [1.0/obs for obs in ratios]
```

```python
fig, ax = plt.subplots()
#plt.hist(ratios_inverse, bins=20, edgecolor="black", color="seagreen")
sns.distplot(ratios_inverse, bins=20, hist_kws={"alpha":0.9, "color":"seagreen"}, kde_kws={"color":'black'})
plt.xlabel('E')
plt.ylabel('Normalised Frequency')
plt.xlim(0, 1)
ax.set_title('Histogram of E', fontsize=20)
plt.show()
```
